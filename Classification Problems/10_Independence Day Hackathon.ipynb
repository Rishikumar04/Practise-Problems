#### Split

```python
train_df
```

```python
X = train_df['TITLE']
y = train_df['Target']
```

```python
test = test_df['TITLE']
```

```python
#train_data
def metrics(y_true,y_pred):
    #print('Confusion Matrix:\n', confusion_matrix(y_true, y_pred))
    print('\n\nAccuracy Score:\n', accuracy_score(y_true, y_pred))
    print('\n\nClassification Report: \n', classification_report(y_true, y_pred))

def vectorizer(vector, X, test):
    vector.fit(X)
    X_vc = vector.transform(X)
    test_vc = vector.transform(test)
    return X_vc, test_vc

def split(vc,y):
    X_train,X_test,y_train, y_test = train_test_split(vc, y, test_size=0.3, random_state=123, stratify=y)
    
    return X_train,X_test,y_train, y_test
    
def predictions(model,vector, X, y, test):
    
    X_vector, test_vector = vectorizer(vector, X, test)
    
    X_train,X_test,y_train, y_test= split(X_vector, y)
    
    model.fit(X_train, y_train)
    
    train_pred = model.predict(X_train)
    test_pred = model.predict(X_test)
    
    actual = [y_train, y_test]
    pred = [train_pred, test_pred]
    text = ('Train', 'Test')
    for i in range(2):
        print(f"-----{text[i]}-----")
        metrics(actual[i], pred[i])
        
#______#
    
    final_pred = model.predict(test_vector)
    
    for index,val in enumerate(final_pred.astype(int)):
        if val==1:
            sample_df.loc[index,'Computer Science'] = 1
        elif val==2:
            sample_df.loc[index,'Physics'] = 1
        elif val == 3:
            sample_df.loc[index,'Mathematics'] = 1 
        elif val == 4:
            sample_df.loc[index,'Statistics'] = 1 
        elif val == 5:
            sample_df.loc[index,'Quantitative Biology'] = 1 
        elif val == 6:
            sample_df.loc[index,'Quantitative Finance'] = 1 
        elif val == 7:
            sample_df.loc[index,['Computer Science', 'Physics']] = 1 
        elif val == 8:
            sample_df.loc[index,['Computer Science', 'Mathematics']] = 1 
        elif val == 9:
            sample_df.loc[index,['Computer Science', 'Statistics']] = 1
        elif val == 10:
            sample_df.loc[index,['Computer Science', 'Quantitative Biology']] = 1
        elif val == 11:
            sample_df.loc[index,['Computer Science', 'Quantitative Finance']] = 1
        elif val == 12:
            sample_df.loc[index,['Physics', 'Mathematics']] = 1
        elif val == 13:
            sample_df.loc[index,['Physics', 'Statistics']] = 1
        elif val == 16:
            sample_df.loc[index,['Mathematics', 'Statistics']] = 1
        elif val == 19:
            sample_df.loc[index,['Statistics', 'Quantitative Biology']] = 1
        elif val == 20:
            sample_df.loc[index,['Statistics', 'Quantitative Finance']] = 1
        elif val == 21:
            sample_df.loc[index,['Quantitative Biology', 'Quantitative Finance']] = 1
        elif val == 22:
            sample_df.loc[index,['Computer Science', 'Physics', 'Mathematics']] = 1
        elif val == 23:
            sample_df.loc[index,['Computer Science', 'Physics', 'Statistics']] = 1
        elif val == 26:
            sample_df.loc[index,['Physics', 'Mathematics', 'Statistics']] = 1
        else:
            print(index, val)
    name = str(model)[:5]+str(vector)[:5]
    sample_df.to_csv(name+'.csv', index = None)
    return final_pred
```
